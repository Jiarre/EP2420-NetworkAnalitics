```python
%matplotlib inline

# Task 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

np.set_printoptions(precision=3)
plt.rcParams.update({'font.size': 22})
```

```python
path_X = "../Project1_data/X.csv"
X = pd.read_csv(path_X)
path_Y = "../Project1_data/Y.csv"
Y = pd.read_csv(path_Y)
```

```python

def LateXprint(D,f):
    mean = np.mean(D[f])
    std = np.std(D[f])
    max = np.max(D[f])
    min = np.min(D[f])
    p25 = np.percentile(D[f],25)
    p50 = np.percentile(D[f],50)
    p90 = np.percentile(D[f],90)
    print(f"{f} & {'{:.2e}'.format(mean)} & {'{:.2e}'.format(std)} & {'{:.2e}'.format(max)} & {'{:.2e}'.format(min)} & {'{:.2e}'.format(p25)} & {'{:.2e}'.format(p50)} & {'{:.2e}'.format(p90)} \\\\")
    
def statsPerFeature(D,f):
    mean = np.mean(D[f])
    std = np.std(D[f])
    max = np.max(D[f])
    min = np.min(D[f])
    p25 = np.percentile(D[f],25)
    p50 = np.percentile(D[f],50)
    p90 = np.percentile(D[f],90)
    print(f"{f},{'{:.2e}'.format(mean)},{'{:.2e}'.format(std)},{'{:.2e}'.format(max)},{'{:.2e}'.format(min)},{'{:.2e}'.format(p25)},{'{:.2e}'.format(p50)},{'{:.2e}'.format(p90)}")
```

```python
print("Feature,Mean,STD,Max,Min,20thP,50thP,90thP" )
for f in X:
    if(f!="TimeStamp"):
        statsPerFeature(X,f)
for f in Y:
    if(f!="TimeStamp"):
        statsPerFeature(Y,f)

np.quantile(X["all_%%usr"],0.25)
```

```python
print("Feature,Mean,STD,Max,Min,20thP,50thP,90thP" )
for f in X:
    if(f!="TimeStamp"):
        LateXprint(X,f)
for f in Y:
    if(f!="TimeStamp"):
        LateXprint(Y,f)
```

```python
### Task 2.1
from sklearn import  linear_model


tmp = X.copy()
del tmp["TimeStamp"]

tmp["DispFrames"] = Y["DispFrames"]
sample_size = 1080
T = tmp.sample(sample_size,replace=False)
tmp = tmp.drop(T.index)
test_labels = T.iloc[:,-1:]
test_features = T.iloc[:,:-1]
train_labels = tmp.iloc[:,-1:]
train_features = tmp.iloc[:,:-1]

regr = linear_model.LinearRegression()
regr.fit(train_features,train_labels)

pred = regr.predict(test_features)
print("Coefficients: ", regr.coef_)
```

```python
def NMAE(gt,pred,test_size):
    psum = 0
    gt = np.array(gt)
    for i in range(0,test_size):
        psum+= abs(gt[i] - pred[i])
    tmp = psum / test_size
    return tmp / np.mean(gt)
pred = regr.predict(test_features)
nmae = NMAE(test_labels,pred,len(test_labels))
naive_pred = np.full(3600,np.mean(train_labels))
naive_nmae = NMAE(test_labels,naive_pred,len(test_labels))
#print(f"NMAE for the model nmae: {'{:.2e}'.format(nmae)}")
#print(f"Naive NMAE: {'{:.2e}'.format(naive_nmae)}")
```
